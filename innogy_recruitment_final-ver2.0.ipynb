```python
#change data for datatype format

#changing type from object to date | I'm not a fan of pandas datetime function

def to_dt(x):
    x = x.split(".") #current data separator
    return datetime.date(int(x[2]),int(x[1]),int(x[0])) #year month day

df["InterviewDate_dt"] = df['InterviewDate'].apply(to_dt)
```

## Supplier's and other brands awarness

```python
cols = ['BA01R_KA_1','BA02R_271_PL','BA02R_272_PL','BA02R_273_PL','BA02R_274_PL','BA02R_279_PL',
        'BA02R_280_PL','BA02R_282_PL','BA02R_283_PL','BA02R_284_PL','BA02R_285_PL','SU01R']

dfs = dfc[cols]

dfs.groupby(['SU01R']).sum()
```

```python
dfs_max= dfs.groupby(['SU01R']).sum().max(axis=1)
dfs_max=dfs_max.rename("max")
```

```python
dfsg = dfs.groupby(['SU01R']).count().join(dfs_max)
dfsg = dfsg[['BA01R_KA_1','max']].rename(columns={"BA01R_KA_1":"count","max":"max_known"}).reset_index()

#dfsg['ratio'] = dfsg.apply(lambda row: row['max_known']/row['count'], axis=1)

dfsg['ratio'] = dfsg['max_known'].div(dfsg['count'])

dfsg
```

### The main idea behind this calculation is that every user lists their own supplier and more.
### More people are supplied by i-th company then total number of people are familiar/aware of that company. Thus some customers do not know who their supplier is (if ratio less then 1). In other words: ratio 97% gives at least 3% of users unaware who are they supplied by.
